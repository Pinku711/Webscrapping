```python
# What is webscrapping:-Web scraping is the process of extracting and parsing data from websites in an automated #
# fashion using a computer program. It's a useful technique for creating datasets for research
# and learning. #

# This module help for internet content.
# setting up the envioronment like !pip install requests and beautifulsoup.and import the element as per your requirement
!pip install requests
! pip install beautifulsoup4
import requests
# parsing .

# Beautifulsoup : Webscrapping library with inbuilt parser .

from bs4 import BeautifulSoup

# Helps to raise a request to the web page .
```

```python
# raising the requests to the web page find.
url="https://quotes.toscrape.com/page/3/"
```

```python
result=requests.get(url) # raising the requests to the web page.
```

```python
result # successful connection to the established, because response 200 means server connection is done with webpage. 
```

```python
data=BeautifulSoup(result.text)
```

```python
print(data)
```

```python
# extrcat the required information .
req_data=data.find_all()
```

```python
req_data
```

```python
heading=[] # empty list 
for i in req_data:
    heading.append(i.text)
```

```python
heading #  Remove unecessary \t--> white space 

for x in range(len(heading)):
    heading[x]=heading[x].strip()
```

```python
heading
```

```python
import pandas as pd
```

```python
h_data=pd.DataFrame(heading)
```

```python
h_data
```

```python
h_data.rename(columns={0:'Headings'},inplace=True)
```

```python
h_data
```

```python
h_data.to_csv('webscrapped_heading_updated.csv')
```
